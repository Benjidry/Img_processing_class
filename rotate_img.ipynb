```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data as data
```

```python
img = data.cat()
```

We are going to use the cat image for this exercicie.

```python
def rotate_image(image, angle, rotation_center):
  row, col, channel = image.shape
  center_x, center_y = rotation_center

  angle_rad = np.deg2rad(angle)

  rotated_image = np.zeros_like(image)

  for y in range(row):
    for x in range(col):
      x_rotated = int((x - center_x) * np.cos(angle_rad) - (y - center_y) * np.sin(angle_rad) + center_x)
      y_rotated = int((x - center_x) * np.sin(angle_rad) + (y - center_y) * np.cos(angle_rad) + center_y)

      if 0 <= x_rotated < col and 0 <= y_rotated < row:
        rotated_image[y, x] = image[y_rotated, x_rotated]

  return rotated_image
```

The "rotate_image" function rotates an image by a specified angle around a given rotation center. It first converts the angle from degrees to radians. It then initializes an empty array for the rotated image. For each pixel in the original image, it calculates its new position after rotation using rotation formulas. If the new position is within the image boundaries, it copies the pixel value to the corresponding location in the rotated image. The function returns the rotated image.

```python
angle = 45
rotation_center = (img.shape[0] // 2, img.shape[1] // 2)
img_rotated = rotate_image(img, angle, rotation_center)
```

```python
plt.imshow(img_rotated)
plt.show()
```
