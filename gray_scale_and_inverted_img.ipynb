```python
img = data.astronaut()
```

Here we obtain an image from the skimage library so we can test with it.

```python
ski.io.imshow(img)
```

We use io.imshow, so we can have a look of the image

```python
img.shape
```

returns a tuple that provides the dimensions of the image. For example, (height, width) for a grayscale image or (height, width, channels) for a color image. This helps in understanding the size and structure of the image data.

```python
img = img[:,:,:3]
img.shape
```

The slicing operation img[:,:,:3] extracts the first three color channels of the image. Here's what this means:

: (first colon) selects all rows (height).
: (second colon) selects all columns (width).
:3 selects the first three channels (RGB) of the image.


```python
imgray_bad = np.sum(img, axis=-1) / 3
imgray_bad = imgray_bad.astype(int)
imgray_bad.shape
```

The code converts an RGB image to an approximate grayscale image by averaging the red, green, and blue color channels. This simplifies the image to a single channel of intensity values, which can be useful for tasks that only require intensity information, such as edge detection or image analysis, where color details are not needed. 

```python
img[100, 100]
```

The expression img[100, 100] retrieves the pixel value at the coordinates (100, 100) from the image array img. This is useful for accessing and examining the color or intensity value of a specific pixel in the image. In an RGB image, img[100, 100] would return an array with three values representing the red, green, and blue color channels at that pixel. 

```python
imgray = np.average(img, axis=-1, weights=[0.299, 0.587, 0.114])
imgray = imgray.astype(int)
```

The code converts an RGB image to a grayscale image using a weighted average of the color channels, where the weights [0.299, 0.587, 0.114]

```python
im = rgb2gray(img)
```


The code im = rgb2gray(img) converts an RGB image to a grayscale image using the rgb2gray function from the skimage.color module.

```python
plt.subplot(131)
plt.imshow(imgray_bad, cmap='gray')
plt.axis('off')
plt.subplot(132)
plt.imshow(imgray, cmap='gray')
plt.axis('off')
plt.subplot(133)
plt.imshow(im, cmap='gray')
plt.axis('off')
plt.show()
```

And we show our images in different gray scales

GRAY SCALE NEGATIVE:

```python
imneg = 255 - im
```

The code imneg = 255 - im creates a negative of a grayscale image. Here's what it does:

Inverts Grayscale Values: This operation subtracts each pixel value in the grayscale image im from 255. Since grayscale values range from 0 to 255, subtracting from 255 effectively inverts the intensity values. For example, a pixel value of 0 (black) becomes 255 (white), and a pixel value of 255 (white) becomes 0 (black).

```python
plt.subplot(121)
plt.imshow(im, cmap='gray')
plt.axis('off')
plt.subplot(122)
plt.imshow(imneg, cmap='gray')
plt.axis('off')
plt.show()
```

And we have our gray image inverted!!!
